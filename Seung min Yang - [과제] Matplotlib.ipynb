```python
# 데이터 읽어 오기. pandas의 read_csv 사용
data = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/data/tips.csv')
data
```

## 1. total_bill 과 tip 의 분포도(scatter) 시각화하기

### tip 금액에 따라 점의 크기 변화 주기, alpha 옵션으로 투명도 지정하기
### 축 label 추가하기, 플롯 title 추가하기


```python
x = data.tip
y = data.total_bill
sizes = x*5

plt.scatter(x, y, s = sizes, alpha=0.4, c = '#138483')
plt.xlabel('Tip', color='white', fontweight='bold')
plt.ylabel('Total_bill', color='white', fontweight='bold')
plt.title('tips.csv', color='yellow', fontweight='bold');
```

## 2. 성별에 따른 tip 비교하기 (bar plat)

```python
x = data.sex
y = data.tip

tips = []
gender = []

for i in y:
  tips = np.append(tips,i)

for i in x:
  gender = np.append(gender,i)

tips_m = []
tips_f = []

for i,j in zip(gender,tips):
  if i == 'Male':
    tips_m = np.append(tips_m,j)
  else:
    tips_f = np.append(tips_f,j)



fig = plt.figure()
fig.subplots_adjust(hspace=0.5, wspace=0.5)
bar_width = 1.5


y_pos_m = np.arange(len(tips_m))
ax1 = fig.add_subplot(211)
ax1.bar(y_pos_m, tips_m, color = '#FF6666', edgecolor='none', width = bar_width)
ax1.xaxis.set_major_formatter(plt.NullFormatter())
plt.ylim(0,10)
plt.ylabel('Male', c = 'white', fontweight='bold')

y_pos_f = np.arange(len(tips_f))
ax2 = fig.add_subplot(212)
ax2.bar(y_pos_f, tips_f, color = '#66FF66', edgecolor='none', width = bar_width)
ax2.xaxis.set_major_formatter(plt.NullFormatter())
plt.xlabel('Tip', c = 'yellow', fontweight='bold')
plt.ylabel('Female', c = 'white', fontweight='bold')
plt.ylim(0,10);
```

## 3. 남자의 여자의 팁 금액의 분포도(histogram)를 그려 비교하세요

### 2 X 1 의 subplot을 이용하세요.

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.5, wspace=0.5)

ax3 = fig.add_subplot(211)
ax3.hist(tips_m, bins=50, color = '#FF6666')
plt.xlim(0,10)
plt.ylabel('Male', c = 'white', fontweight='bold')

ax4 = fig.add_subplot(212)
ax4.hist(tips_f, bins=50, color = '#66FF66')
plt.xlim(0,10)
plt.ylabel('Female', c = 'white', fontweight='bold')
plt.xlabel('Tip', c = 'yellow', fontweight='bold');
```
